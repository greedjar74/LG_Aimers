```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston

"""
1. ì‚¬ì´í‚·ëŸ°ì— ì¡´ì¬í•˜ëŠ” ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¤ê³ , 
   ë°ì´í„°ì™€ ë³€ìˆ˜ ì´ë¦„ì„ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ë¥¼ êµ¬í˜„í•©ë‹ˆë‹¤.
   
   Step01. ì‚¬ì´í‚·ëŸ°ì— ì¡´ì¬í•˜ëŠ” boston ë°ì´í„°ë¥¼ 
           (X, y)ì˜ í˜•íƒœë¡œ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤. 
   
   Step02. ë°ì´í„°ì˜ ë³€ìˆ˜ ì´ë¦„ì„ feature_names ì— ì €ì¥í•©ë‹ˆë‹¤.
   
   [ì‹¤ìŠµ6]ì—ì„œ êµ¬í˜„í•œ í•¨ìˆ˜ë¥¼ ê·¸ëŒ€ë¡œ ì´ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
"""
def load_data():
    
    X, y = load_boston(return_X_y=True)
    
    feature_names = load_boston().feature_names
    
    return X,y,feature_names
    
"""
2. ì—˜ë¼ìŠ¤í‹± ë„·(ElasticNet) íšŒê·€ë¥¼ êµ¬í˜„í•˜ê³ , 
   í•™ìŠµìš© ë°ì´í„°ë¥¼ ë°”íƒ•ìœ¼ë¡œ í•™ìŠµì‹œí‚¨ ëª¨ë¸ì„ 
   ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì™„ì„±í•©ë‹ˆë‹¤.
   
   Step01. ì‚¬ì´í‚·ëŸ°ì— êµ¬í˜„ë˜ì–´ ìˆëŠ” 
           ì—˜ë¼ìŠ¤í‹± ë„·(ElasticNet) íšŒê·€ ëª¨ë¸ì„ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
           
           íŒŒë¼ë¯¸í„° alphaì™€ l1_ratioë¥¼ ììœ ë¡­ê²Œ ì„¤ì •í•©ë‹ˆë‹¤.
   
   Step02. ë¶ˆëŸ¬ì˜¨ ëª¨ë¸ì„ í•™ìŠµìš© ë°ì´í„°ì— ë§ì¶° í•™ìŠµì‹œí‚µë‹ˆë‹¤.
"""

def ElasticNet_regression(train_X, train_y):
    
    ElasticNet_reg = ElasticNet(alpha=0.001, l1_ratio=0.001)
    
    ElasticNet_reg.fit(train_X, train_y)
    
    return ElasticNet_reg
    
    
# ê° ë³€ìˆ˜ì˜ beta_i í¬ê¸°ë¥¼ ì‹œê°í™”í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.
def plot_graph(coef):
    coef.plot(kind='bar')
    
    plt.savefig("result.png")
    elice_utils.send_image("result.png")
    
    
def main():
    
    X,y,feature_names = load_data()
    
    train_X, test_X, train_y, test_y = train_test_split(X,y,test_size=0.2, random_state=100)
    
    elasticnet_reg = ElasticNet_regression(train_X, train_y)
    
    # ì—˜ë¼ìŠ¤í‹±ë„· íšŒê·€ ëª¨ë¸ í‰ê°€ ì ìˆ˜ ì¶œë ¥í•˜ê¸°
    score = elasticnet_reg.score(test_X,test_y)
    print("ElasticNet íšŒê·€ì˜ í‰ê°€ ì ìˆ˜:", score)
    
    # ì—˜ë¼ìŠ¤í‹±ë„· íšŒê·€ì˜ beta_iì˜ í¬ê¸°ë¥¼ ì €ì¥í•©ë‹ˆë‹¤.
    ElasticNet_coef = pd.Series(elasticnet_reg.coef_, feature_names).sort_values()
    print("\nElasticNet íšŒê·€ì˜ beta_i\n", ElasticNet_coef)
    
    plot_graph(ElasticNet_coef)
```
