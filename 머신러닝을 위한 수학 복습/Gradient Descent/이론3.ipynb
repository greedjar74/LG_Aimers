# Gradient Descent 실습

## 점들의 class를 구분하는 boundary 함수를 구한다.
- classification model
- 손 계산

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
n_pts = 500

# x = [(x1, y1), (x2, y2), ... ] -> 이런 쌍이 500개 -> 점이 500개
# y = [0, 1, 1, 0, 0, ... ] -> 각 점들의 class가 나열되어있다.
X, y = datasets.make_circles(n_samples=n_pts, random_state=123, noise=0.2, factor=0.3)

x_data = np.array(X) # X값은 ndarray 형태로
y_data = np.array(y.reshape(500, 1)) # y 값을 ndarray 형태로 -> 500 by 1 행렬
```

```python
def scatter_plot():
    plt.scatter(X[y==0, 0], X[y==0, 1], color='r') # X의 점들 중 class가 0인 점들을 빨간색으로 표시
    plt.scatter(X[y==1, 0], X[y==1, 1], color='b') # X의 점들 중 class가 1인 점들을 파란색으로 표시
    
scatter_plot()
```

# 예시문제

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = 0.7 * x
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(x, y)
plt.show()
```

```python
# 예측과 실제 그래프 비교
def compare_pred(x, pred, y):
    plt.figure(figsize=(7, 7))
    plt.scatter(x, y, label='ground-truth')
    plt.scatter(x, pred, label='prediction')
    plt.legend()
    plt.show()
```

```python
# linear regression model : f(x) = Wx + b
W = np.random.uniform(0, 1)
b = np.random.uniform(0, 1)

# 학습 전 그래프
pred = W*x + b
compare_pred(x, pred, y)

learning_rate = 0.5 # 실제 모델을 만들 때는 0.001정도를 준다.
Epochs = 100 # 학습 반복 횟수

# Gradient Descent
for epoch in range(Epochs):
    y_pred = W*x + b
    
    objective = np.abs(y_pred - y).mean() # Mean Squared Error -> 현재는 그냥 절댓값 사용
    
    # cost함수 미분
    w_grad = ((y_pred - y) * x).mean()
    b_grad = (y_pred - y).mean()
    
    # 가중치 업데이트
    W = W - learning_rate * w_grad
    b = b - learning_rate * b_grad
    
# 학습 후 그래프
pred = W*x + b
compare_pred(x, pred, y)
```
